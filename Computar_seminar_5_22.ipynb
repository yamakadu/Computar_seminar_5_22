## markdownとは
markdownは文章を記述するための言語の一つ。つまり、書き方のこと。





## markdown記法のメリット



### 1. 他サイトからコピペしてもフォントや文字の大きさがおかしくならない
他サイトから文字をコピペして引用しても、文字の大きさやフォント、色等が変にならない
→変更する場合はhtmlタグを用いる。
(タグというマークを使って、HTMLの文書構造を指定したり、文字の大きさを変えたり、、、)



### 2. HTMLタグが使える
そのままTwitterやYoutubeの埋め込みコードを貼り付けることが可能。
普通にhtmlで描く場合と比較すると圧倒的に描きやすい。htmlに簡単に変換できる



### 3. どこでも記事を書ける
　特定のウェブサイトを使わなくてはならない書き方ではないので、いつでもどこでもオンラインでもオフラインでも書ける。
 (Evernoteなど使ってたら、描くのが非常に楽になる)
 
 
 
 ### 4. 汎用性が高い
Githubやjyupter notebookでも利用されている書き方なので、ブログだけでなく他のウェブサイトでも使える


## markdown記法のメリット


### 1. 書き方を覚えるのが大変
覚えるまでに時間がかかる。10回くらい記事を書けばなれるらしい


### 2. 文法間違いで地味に時間を消費してしまう
多少書き方がややこしいところはある

## 見出し
見出しの先頭部分に#をつける。#が増えるごとに小さな見出しとなる

# 大きい見出し
## 中くらい見出し
### 小さい見出し
#### さらに小さい見出し

## リスト
リストには通常のリストと数字つきリストがある
まず、数字付きリストは先頭部分に1.とつけると表示できる。(数字部分は他の数字でも可)

1. 数字付きリスト1
1. 数字付きリスト2
1. 数字付きリスト3 


通常のリストは、先頭部分に-,+または*のどれかをつけると表示できる。
また、リストではtabキーを使うことにより階層化となる。



- 通常リスト
* 通常リスト
+ 通常リスト

- 階層化１
    - 階層化2-1
    - 階層化2-2 

## 見出し
区切り線を表示するには、*か-か_のどれかを３つ並べて入力する

***
---
___

## 強調
文章の一部など、強調したい部分を**で囲うと強調された文字になる

**この場所**を強調する

## コードの表示
コードの表示では、バッククォートを使用して表示を行う。

tml、css、php、ruby、python…ほぼなんでも言語に対応している

```
 ここにコードを記入する
 ```

言語指定したい場合は始めのバッククオートに文字を追加する


```c
#include <stdio.h>

int main() {
   
    
    printf("Hello World");
    
    
};
```


出力した値を反映させる時はsh


```sh
100 200
300 400
```

## リンク

[リンク名](URL)

画像の表示は、


![代替テキスト](画像のURL)

によって行える(ドラッグアンドドロップや画像挿入でももちろん可)
